## 编程题目：财经快讯智能体协作

### 题目背景

在量化投资和金融数据分析中，财经新闻对上市公司股价的影响至关重要。请你设计并实现两个智能体（Agent），协作完成财经快讯的采集、结构化和影响的自动抽取。

大模型根据自己电脑配置部署相匹配的版本，建议优先使用qwen系列。

可以使用已有的多智能体框架。


---

### 任务要求

#### 智能体A：上市公司新闻采集与结构化

1. **数据采集**  
   从东方财富网财经快讯（[https://kuaixun.eastmoney.com/](https://kuaixun.eastmoney.com/)）抓取**与A股上市公司相关**的最新财经新闻。

   ![Alt text](image.png)


2. **结构化处理**  
   对每条新闻，提取以下字段：  
   - 新闻时间  
   - 新闻标题  
   - 新闻正文  
   - 涉及公司（单个公司名）
   - 股票代码（单个公司代码）
   - 股票简称（单个公司简称）
3. **本地保存**  
   **如果一条新闻涉及多家公司，则为每个公司分别保存一行，其他字段内容相同。**  
   将结构化后的新闻数据保存为本地CSV文件（如`news.csv`），字段顺序如下：  
   ```
   时间, 标题, 正文, 涉及公司, 股票代码, 股票简称
   ```

#### 智能体B：新闻影响因子抽取

1. **读取数据**  
   读取智能体A保存的CSV财经新闻数据。
2. **大模型智能分析**  
   对每条新闻，针对每个公司，使用大模型：
   - 生成该新闻的简明摘要（新闻摘要）。
   - 判断该新闻对该上市公司的影响方向，输出为：  
     - +1（正面）  
     - -1（负面）  
     - 0（中性）
3. **因子表生成**  
   生成一个因子表，保存为本地CSV文件（如`factor.csv`），字段顺序如下：  
   ```
   公司名, 股票代码, 股票简称, 新闻时间, 新闻标题, 影响方向, 新闻摘要
   ```

---

### 多公司新闻的处理要求

- **当一条新闻涉及多家上市公司时，news.csv中为每个公司分别保存一行，其他字段内容相同。**
- factor.csv同理，每行只对应一个公司。

- **A股公司名、股票代码、股票简称**可通过东方财富网或公开A股公司列表获取对照表。
- **新闻摘要**和**影响方向**均需通过大模型（如qwen3）对新闻正文和公司关系智能生成，不得简单截取或用关键词规则。
- 代码需结构清晰，注释完善。
